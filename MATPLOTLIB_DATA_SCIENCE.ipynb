```python
# Matplotlib

# Data Visualization
# Matplotlib
# Types pf Plot
# Generating GraphsFrom CSV

import matplotlib.pyplot as plt
```

```python
# plot([x-axis],[y-axis])
plt.plot([1,2,3],[4,5,1])

# show() --> used to display graph
plt.show()
```

```python
# add title and label to graph
x = [5,8,10]
y = [12,16,6]
plt.plot(x,y)
plt.title('Info')
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')
plt.show()
```

```python
# Add stylist to a graph
from matplotlib import style
style.use('ggplot')  # can use other type too, search on internet (customizing matplot with style)
x = [6,12,15]
y = [12,16,6]
x2 = [12,15,19]
y2 = [13,15,10]
plt.plot(x,y,'g',label='Line One',linewidth=5)  # g for green colour
plt.plot(x2,y2,'c',label='Line Two',linewidth=5)
plt.title('Info')
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')
plt.legend()
plt.grid(True,color='k')  # k --> black
plt.show()
```

```python
# Generate Bar Graph
plt.bar([1,6,7,9,10],[1,3,6,9,12],label='Example One')
plt.bar([2,5,6,8,12],[3,6,7,13,15],label='Example Two')
plt.legend()
plt.title('Info')
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')
plt.show()
```

```python
# Scatter plot
x = [1,2,3,4,5,6,7]
y = [2,4,6,7,8,9,10]
plt.scatter(x,y,label='scatter',color='g',s=100,marker='o')  # s --> size of marker   # m--> 'shape'
plt.title('Scatter Graph')
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')
plt.show()
```

```python
# Pie plot
slices = [7,3,2,4,5]
activities = ['Sleeping','Working','Relax','Play','Eating']
cols = ['c','m','r','y','k']
plt.pie(slices,
        labels=activities,
        colors=cols,
        startangle = 90,
        shadow=True,
        explode =(0.1,0,0,0,0),
        autopct = '%1.1f%%')  # overlay percentage over graph
plt.show()
```

```python
# How to change the label position
from matplotlib import style
style.use('ggplot')  # can use other type too, search on internet (customizing matplot with style)
x = [6,12,15]
y = [12,16,6]
x2 = [12,15,19]
y2 = [13,15,10]
plt.plot(x,y,'g',label='Line One',linewidth=5)  # g for green colour
plt.plot(x2,y2,'c',label='Line Two',linewidth=5)
plt.title('Info')
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')
plt.legend(loc=3)     ## Search in google for the code (matplotlib pyplot legend)
plt.grid(True,color='k')  # k --> black
plt.show()
```
